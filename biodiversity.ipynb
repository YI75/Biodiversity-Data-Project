```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
observation_df = pd.read_csv('observations.csv')

observation_df.head(10)
```

```python
observation_df.shape
```

```python
species_df = pd.read_csv('species_info.csv')

species_df.head(10)
```

```python
species_df.shape
```

```python
list(set(observation_df.columns) & set(species_df.columns))
```

```python
bio_diversity = pd.merge(observation_df, species_df, on='scientific_name')
bio_diversity.head(50)
```

```python
bio_diversity.info()
```

```python
bio_diversity.duplicated().value_counts()
```

```python
bio_diversity.isnull().value_counts()
```

```python
bio_diversity['category'].value_counts(normalize=True)
```

```python
category = bio_diversity.groupby('category')['observations'].sum()
category
```

```python
plt.figure(figsize=(12,5))
sns.set_style("whitegrid")

ax1 = plt.subplot(1,2,1)
sns.barplot(category[['Nonvascular Plant', 'Vascular Plant']], palette='deep')
plt.title('Number of Observations for Plants')
plt.xlabel('Type of Plant')
plt.ylabel('Count')

ax2 = plt.subplot(1,2,2)
sns.barplot(category[['Amphibian', 'Bird', 'Fish', 'Mammal', 'Reptile']], palette='deep')
plt.title('Number of Observations for Animals')
plt.xlabel('Type of Animal')
plt.ylabel('Count')

plt.subplots_adjust(wspace = 0.3)
```
