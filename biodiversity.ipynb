```python
bio_diversity = pd.merge(observation_df, species_df, on='scientific_name')
bio_diversity.head(50)
```

```python
bio_diversity['category'].value_counts(normalize=True)
```

```python
category = bio_diversity.groupby('category')['observations'].sum()
category
```

```python
plt.figure(figsize=(12,5))
sns.set_style("whitegrid")

ax1 = plt.subplot(1,2,1)
sns.barplot(category[['Nonvascular Plant', 'Vascular Plant']], palette='deep')
plt.title('Number of Observations for Plants')
plt.xlabel('Type of Plant')
plt.ylabel('Count')

ax2 = plt.subplot(1,2,2)
sns.barplot(category[['Amphibian', 'Bird', 'Fish', 'Mammal', 'Reptile']], palette='deep')
plt.title('Number of Observations for Animals')
plt.xlabel('Type of Animal')
plt.ylabel('Count')

plt.subplots_adjust(wspace = 0.3)
```

```python
bio_diversity['conservation_status'].isnull().value_counts(normalize=True)
```

```python
bio_diversity.fillna('Unknown', inplace=True)
bio_diversity['conservation_status'].isnull().value_counts(normalize=True)
```

```python
category_status_df = bio_diversity.groupby(['category','conservation_status'])['observations'].sum()
category_status_df = category_status_df.reset_index()
category_status_df.head(15)
```

```python
category_status_df = category_status_df.pivot_table(index="category",columns="conservation_status",values="observations",
    aggfunc="sum",
    fill_value=0
).reset_index()
category_status_df
```

```python
animals = ['Amphibian', 'Bird', 'Fish', 'Mammal', 'Reptile']
plants = ['Nonvascular Plant', 'Vascular Plant']
no_unknowm = ['category','Endangered','In Recovery','Species of Concern','Threatened']

animals_df = category_status_df[category_status_df['category'].isin(animals)]
plants_df = category_status_df[category_status_df['category'].isin(plants)]

animals_df.plot(x= 'category', kind = 'bar', stacked= True, title='Stacked Bar Graph by Animals of Conversation Status')
```

```python
animals_df[no_unknowm].plot(x= 'category', kind = 'bar', stacked= True, title='Stacked Bar Graph by Animals of Conversation Status')
```

```python
plants_df.plot(x= 'category', kind = 'bar', stacked= True, title='Stacked Bar Graph by Plants of Conversation Status')
```

```python
plants_df[no_unknowm].plot(x= 'category', kind = 'bar', stacked= True, title='Stacked Bar Graph by Plants of Conversation Status')
```
